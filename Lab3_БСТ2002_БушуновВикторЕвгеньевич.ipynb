```python
def prefix(b):
    pr = [0] * len(b)
    for i in range(1, len(b)):
        x = pr[i - 1]
        while x > 0 and b[x] != b[i]:
            x = pr[x - 1]
        if b[x] == b[i]:
            x = x + 1
        pr[i] = x
    return pr

# aabaababaabaabaaab
#
# aabaabaa
# 01012345 - f(x)

def kmp(a, b):

    a = a.lower()
    b = b.lower()

    ind = -1
    pr = prefix(b)
    x = 0
    for i in range(len(a)):
        while x > 0 and b[x] != a[i]:
            x = pr[x - 1]
        if b[x] == a[i]:
            x = x + 1
        if x == len(b):
            ind = i - len(b) + 1
            break
    return ind


a = input("Введите исходную строку: ")
b = input("Введите строку для поиска: ")

print("Номер элемента начала искомой строки равен: ", kmp(a, b))
```

```python
def table(a):
    dictionary = dict()
    for i in range(len(a) - 1):
        dictionary[a[i]] = len(a) - i - 1
    if not a[len(a) - 1] in dictionary:
        dictionary[a[len(a) - 1]] = len(a)
    return dictionary

# Hoola-Hoola girls Like Hooligans
#                        Hooligan

def mur(a, b):

    a = a.lower()
    b = b.lower()

    dictionary = table(b)
    currStep = len(b) - 1

    while currStep < len(a):
        indB = len(b) - 1

        for i in range(currStep, currStep - len(b), -1):
            if a[i] != b[indB]:
                if a[i] in dictionary:
                    step = dictionary[a[i]]
                else:
                    step = len(b)
                break

            if i == currStep - len(b) + 1:
                return i
            indB = indB - 1

        currStep = currStep + step


a = input("Введите исходную строку: ")
b = input("Введите строку для поиска: ")

print("Номер элемента начала искомой строки равен: ", mur(a, b))
```
